## Step1 - Introduction -> Give a detailed data description and objective

**Dataset Description:**

The Aspiring Mind Employment Outcome 2015 (AMEO) dataset, released by Aspiring Minds, focuses on employment outcomes for engineering graduates. It includes dependent variables such as Salary, Job Titles, and Job Locations, along with standardized scores in cognitive skills, technical skills, and personality skills. With around 40 independent variables and 4000 data points, these variables encompass both continuous and categorical data. The dataset also includes demographic features and unique identifiers for each candidate.

**Objective:**

The goal of this Exploratory Data Analysis (EDA) is to extensively investigate the provided dataset, with a particular emphasis on understanding the link between various variables and the target variable, Salary.

The key aims of this analysis include:
- Providing a detailed explanation of the dataset's features.
- Find any observable patterns or trends in the data.
- Investigating the relationships between the independent factors and the target variable (salary).
- Identify any outliers or abnormalities in the dataset.
- Offering practical insights and recommendations based on the analysis.

#### Finding Outliers using IQR

```python
for column in numerical_df.select_dtypes(include=['number']):
    
    Q1 = numerical_df[column].quantile(0.25)
    
    Q3 = numerical_df[column].quantile(0.75)
    
    IQR = Q3 - Q1
    
    # Find outliers
    outliers = numerical_df[(numerical_df[column] < (Q1 - 1.5 * IQR)) | (numerical_df[column] > (Q3 + 1.5 * IQR))]
    
    # Print outliers
    print("Outliers in column", column, ":", outliers)
```

```python
numerical_columns = numerical_df.select_dtypes(include=['int', 'float']).columns

for column in numerical_columns:
    # Calculate the frequency distribution
    frequency_distribution = numerical_df[column].value_counts(normalize=True)

    # Plot the KDE for the current numerical column
    plt.figure(figsize=(4, 4))
    sns.kdeplot(numerical_df[column], shade=True)
    plt.title(f'Kernel Density Estimation of {column}')
    plt.xlabel(column)
    plt.ylabel('Density')
    plt.show()
```

```python
from scipy import stats

numerical_columns = numerical_df.select_dtypes(include=['int', 'float']).columns

for column in numerical_columns:
    # Calculate the probability distribution
    probability_distribution = numerical_df[column].value_counts(normalize=True)
    
    # Plot the probability distribution using probplot
    plt.figure(figsize=(5, 5))
    stats.probplot(numerical_df[column], dist="norm", plot=plt)
    plt.title(f'Probability Distribution of {column}')
    plt.xlabel('Theoretical quantiles')
    plt.ylabel('Ordered Values')
    plt.show()
```

#### Step5 - Research Questions
- Times of India article dated Jan 18, 2019 states that “After doing your Computer Science Engineering if you take up jobs as a Programming Analyst, Software Engineer, Hardware Engineer and Associate Engineer you can earn up to 2.5-3 lakhs as a fresh graduate.” Test this claim with the data given to you.
- Is there a relationship between gender and specialization? (i.e. Does the preference of Specialisation depend on the Gender?)

```python
relevant_roles = ['Programming Analyst', 'Software Engineer', 'Hardware Engineer', 'Associate Engineer']
fresh_graduates_data = df[df['Designation'].isin(relevant_roles) & (df['Experience'] == 'Fresh Graduate')]

summary_statistics = fresh_graduates_data['Salary'].describe()

claimed_salary_range = (250000, 300000)
is_claim_verified = (summary_statistics['min'] >= claimed_salary_range[0]) and (summary_statistics['max'] <= claimed_salary_range[1])

print("Summary Statistics for Fresh Graduates in Relevant Job Roles:")
print(summary_statistics)
if is_claim_verified:
    print("The claimed salary range is verified.")
else:
    print("The claimed salary range is not verified.")
```

```python
tab = pd.crosstab(df['Gender'], df['Specialization'], normalize='index')

ax = tab.plot(kind='bar', figsize=(12, 6))

plt.show()
```
