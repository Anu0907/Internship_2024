# **Importing Packages**

```python
import pandas as pd
import numpy as np
```

# **Loading Dataset**

```python
df1 = pd.read_csv('/content/drive/MyDrive/Innomatics/Assignments/diminos_data_v2/deliveries.csv')
df2 = pd.read_csv('/content/drive/MyDrive/Innomatics/Assignments/diminos_data_v2/orders.csv')
df3 = pd.read_csv('/content/drive/MyDrive/Innomatics/Assignments/diminos_data_v2/order_items.csv')

merged_df = pd.merge(df1, df2, on='order_id')
merged_df = pd.merge(merged_df, df3, on='order_id')
merged_df.head()
```

```python
merged_df.info()
```

```python
merged_df.shape
```

```python
merged_df['time_stamp'] = pd.to_datetime(merged_df['time_stamp'])

merged_df['time_status'] = merged_df['time_stamp'].dt.time

merged_df['min_status'] = merged_df['time_stamp'].dt.hour * 60 + merged_df['time_stamp'].dt.minute + merged_df['time_stamp'].dt.second / 60

merged_df['order_placed_at'] = pd.to_datetime(merged_df['order_placed_at'])

merged_df['time_placed'] = merged_df['order_placed_at'].dt.time

merged_df['min_placed'] = merged_df['order_placed_at'].dt.hour * 60 + merged_df['order_placed_at'].dt.minute + merged_df['order_placed_at'].dt.second / 60

merged_df.head()
```

```python
df_new = merged_df.drop(['time_status','time_placed'],axis =1)
df_new.head()
```

```python
df_new.info()
```

```python
df_new['actual_min'] = df_new['min_status'] - df_new['min_placed']
df_new.head()
```

```python
df_new['items_ordered'] = df_new['items_ordered'].astype(str)
df_new['items_ordered'] = df_new['items_ordered'].str.strip("[]").str.replace("'", "").str.split(', ')

df_expanded = df_new['items_ordered'].apply(pd.Series)
df_expanded.columns = ['p1', 'p2', 'p3']
```

```python
df_new = pd.merge(df_new, df_expanded, left_index=True, right_index=True, how='inner')
df_new.head()
```

## **What is the total amount Pizza store lost in 2023 due to refunds on late deliveries?**

```python
start_date = pd.to_datetime('2023-01-01')
end_date = pd.to_datetime('2023-12-31')

df_new_1 = df_new[(df_new['order_placed_at'] >= start_date) & (df_new['order_placed_at'] <= end_date)]

print(df_new_1.shape)
df_new_1.head()
```

```python
df_new_1["status"].value_counts()
```

```python
lost_df = df_new_1[df_new_1["status"] == 'Delivered']

lost_df["status"].value_counts()
```

```python
lost_df = lost_df[lost_df["actual_min"] > 30]

print(lost_df.shape)
lost_df.head()
```

```python
lost_df[['p1', 'p2','p3']] = lost_df[['p1', 'p2','p3']].astype(int)

lost_df.info()
```

```python
p1 = lost_df["p1"].sum()
p2 = lost_df["p2"].sum()
p3 = lost_df["p3"].sum()

total = p1 + p2 + p3

print("Total amount by store lost due to late delivery in 2023 is:",total)
```
