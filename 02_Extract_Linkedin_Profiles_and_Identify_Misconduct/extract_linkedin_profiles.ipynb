```python
import os
import re
```

```python
all_chats = ""

for root, dirs, files in os.walk('data'):
    for file in files:
        if file.endswith('.txt'):
            file_path = os.path.abspath(os.path.join(root, file))
            with open(file_path, 'r', encoding='latin-1') as f:
                all_chats += f.read()
                
print(all_chats[:1000])
```

```python
all_chats = all_chats.lower()
```

### 1. How many LinkedIn URLs were there over all the chat files?

```python
pattern = "linkedin.com"

all_linkedin_urls = re.findall(pattern, all_chats)

print(len(all_linkedin_urls))
```

### 2. Out of all the LinkedIn URLs shared over chat, how many are valid profile URLs?

```python
pattern = r"linkedin.com/in/[\w-]+"

linkedin_profile_urls = re.findall(pattern, all_chats)

print(len(linkedin_profile_urls))
```

### 3. How many valid LinkedIn Profile URLs are left after duplicate removal?

```python
print(len(set(linkedin_profile_urls)))
```

### 4. How many unique occurrences of the misconduct were detected?

```python
pattern = r"us02web.zoom.us[^\s]+"

zoom_urls = re.findall(pattern, all_chats)

print(len(set(zoom_urls)))
```
